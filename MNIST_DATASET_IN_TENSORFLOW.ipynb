```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import tensorflow
from tensorflow.examples.tutorials.mnist import input_data
```

```python
mnist = input_data.read_data_sets("MNIST_data/", one_hot=True)
```

```python
# training set
train_images = mnist.train.images
train_labels = mnist.train.labels
# validation set
validation_images = mnist.validation.images
validation_labels = mnist.validation.labels
# test set
test_images = mnist.test.images
test_labels = mnist.test.labels
```

```python
# print the number of data points in each dataset
print(mnist.train.num_examples)
print(mnist.validation.num_examples)
print(mnist.test.num_examples)
```

```python
print(mnist.train.images.shape)
im = mnist.train.images[0]
print(im.shape)
# The images in Mnist dataset are all normalized from 28 by 28 matrix to one-dimensional matrix, which shape is (784,0).
```

```python
print(np.min(im),np.max(im))
# Unnormalized data ranges from 0 to 255.
```

```python
im = im.reshape(28, 28)
print(im.shape)
plt.imshow(im, cmap='Greys')
plt.show()
```

```python
print(mnist.train.labels.shape)
print(mnist.train.labels[0])
# The MNIST dataset is labeled by a number ranging from 0 to 9. An one-hot vector consists of 0 and 1, where it has only a number of 1. 
```

```python
indexs = np.where(np.argmax(train_labels, axis=1) == 7)
print(indexs)
fig, ax = plt.subplots(
    nrows=5,
    ncols=5,
    sharex='all',
    sharey='all', )

ax = ax.flatten()
for i in range(25):
    img = train_images[indexs[0][i]].reshape(28, 28)
    ax[i].imshow(img, cmap='Greys', interpolation='nearest')

ax[0].set_xticks([])
ax[0].set_yticks([])
plt.tight_layout()
plt.show()
```

```python
X = []
Y = []

for i in range(10):
    x = i
    y = np.sum(np.argmax(train_labels, axis=1) == i)
    X.append(x)
    Y.append(y)
    plt.text(x, y, '%s' % y, ha='center', va= 'bottom')

plt.bar(X, Y, facecolor='#9999ff', edgecolor='white')
plt.xticks(X)
plt.show()
```
